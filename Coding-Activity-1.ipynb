```python
import numpy as np
import matplotlib.pyplot as plt

def x_t(t):
    return 9 * np.cos(200 * np.pi * t + 0.4 * np.pi)

# -0.02 <= t <= 0.02
t_min = -0.02
t_max = 0.02
t_continuous = np.linspace(t_min, t_max, 1000)  # for line tracing

sampling_intervals = [0.0075, 0.005, 0.0005]
sampling_labels = ['T_s = 0.0075 s', 'T_s = 0.005 s', 'T_s = 0.0005 s']

# Create figure
plt.figure(figsize=(10, 8))

# Colors for different sample sets
colors = ['b', 'g', 'r']  # Blue, Green, Red for different plots

for i, Ts in enumerate(sampling_intervals):
    t_values = np.arange(t_min, t_max, Ts)
    x_values = x_t(t_values)
    x_continuous = x_t(t_continuous)
    
    plt.subplot(3, 1, i + 1)
    plt.plot(t_continuous, x_continuous, color=colors[i], label="Continuous Trace")  # Color for continuous line
    plt.stem(t_values, x_values, linefmt=f'{colors[i]}-', markerfmt=f'{colors[i]}o', basefmt=" ", label="Sampled Points")  # Color for stems and points
    plt.title(f'Sampled Signal with {sampling_labels[i]}')
    plt.xlabel('Time (s)')
    plt.ylabel('x(t)')
    plt.legend()

plt.tight_layout()
plt.savefig('Coding-Activity-1-Image.png')
```
